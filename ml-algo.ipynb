```python
df_country=pd.read_excel('Zomatodataset/Country-Code.xlsx')
df_country.head()
```

```python
df_country.isna().sum()
```

### Joining Dataframes

```python
final_df=df.merge(df_country,how='left',on='Country Code')
final_df.head(2)
```

```python
final_df.dtypes
```

## Counting the existance of the values in dataframe

```python
final_df.Country.value_counts()
country_name=final_df.Country.value_counts().index
country_count=final_df.Country.value_counts().values
```

```python
plt.pie(country_count[:3],labels=country_name[:3],autopct='%1.2f%%')
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
ratings
```

```python
#Observations
import matplotlib
matplotlib.rcParams['figure.figsize']=(20,10)
sns.barplot(x="Aggregate rating",y="Rating Count",data=ratings)
```

```python
countries_with_ratings=final_df[final_df['Aggregate rating']==0].groupby(['Aggregate rating','Country']).size().reset_index()
countries_with_ratings
```

```python
country_currency=final_df[['Currency','Country']].groupby(['Country','Currency']).size().reset_index()
country_currency
```

```python
coutries_with_online_delivery=final_df[final_df['Has Online delivery']=='Yes'].groupby(['Country']).size().reset_index()
coutries_with_online_delivery
```
